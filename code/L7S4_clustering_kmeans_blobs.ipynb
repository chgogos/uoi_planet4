```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

plt.rcParams["figure.figsize"] = [12, 7]

X, y = make_blobs(
    n_samples=200,
    n_features=2,
    centers=4,
    cluster_std=1.5,
    shuffle=True,
    random_state=42,
)

plt.scatter(X[:, 0], X[:, 1], edgecolor="black", s=50)
plt.show()
```

```python
from sklearn.cluster import KMeans

km = KMeans(
    n_clusters=4, init="random", n_init=10, max_iter=1000, tol=1e-04, random_state=42
)
y_km = km.fit_predict(X)
```

```python
import matplotlib.pyplot as plt

plt.scatter(X[:, 0], X[:, 1], c=km.labels_, edgecolor="black")
plt.show()
```

```python
# plot clusters
plt.scatter(
    X[y_km == 0, 0], X[y_km == 0, 1],
    s=50, c='lightgreen',
    marker='s',edgecolor='black',
    label='cluster 1'
)

plt.scatter(
    X[y_km == 1, 0], X[y_km == 1, 1],
    s=50, c='orange',
    marker='o', edgecolor='black',
    label='cluster 2'
)

plt.scatter(
    X[y_km == 2, 0], X[y_km == 2, 1],
    s=50, c='lightblue',
    marker='v', edgecolor='black',
    label='cluster 3'
)

plt.scatter(
    X[y_km == 3, 0], X[y_km == 3, 1],
    s=50, c='pink',
    marker='^', edgecolor='black',
    label='cluster 4'
)

# plot the centroids
plt.scatter(
    km.cluster_centers_[:, 0], km.cluster_centers_[:, 1],
    s=500, marker='*',
    c='red', edgecolor='black',
    label='centroids'
)

plt.legend(scatterpoints=1)
plt.grid()
plt.show()
```
