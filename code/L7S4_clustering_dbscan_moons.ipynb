```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons

plt.rcParams["figure.figsize"] = [12, 7]

X, y = make_moons(n_samples=300, noise=0.05, random_state=42)

plt.scatter(X[:, 0], X[:, 1], edgecolor="black", s=50)
plt.show()
```

```python
import numpy as np
from sklearn.cluster import DBSCAN

dbs = DBSCAN(eps=0.15, min_samples=10)
dbs.fit(X)

no_clusters = len(np.unique(dbs.labels_))
if -1 in dbs.labels_:
    no_clusters -= 1
no_noise_points = len([x for x in dbs.labels_ if x == -1])
print(f"Number of clusters = {no_clusters}")
print(f"Number of noise points = {no_noise_points}")
```

```python
import matplotlib.pyplot as plt

plt.scatter(X[:, 0], X[:, 1], c=dbs.labels_, edgecolor="black", s=50)
plt.title(f"{no_clusters} clusters with data")
plt.show()
```
