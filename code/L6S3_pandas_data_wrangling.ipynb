```python
import pandas as pd
```

```python
# Read html table with Pandas
url = 'https://en.wikipedia.org/wiki/List_of_countries_by_number_of_births'
tables = pd.read_html(url)
print(f"Number of tables in this web page: {len(tables)}")
```

```python
# get as dataframe the first and only table of the returned tables
df = tables[0]
# save data for offline access
df.to_csv('../data/births2021.csv',  index=False)
df
```

```python
# drop column Rank, rename columns
df2 = df.drop('Rank', axis=1)
df2.rename({'Country':'country', 'Number of births (2021)':'births_2021'}, axis=1, inplace=True)
df2
```

```python
# find countries that have births +-10% of births in Greece for 2021
births_in_greece = df2[df2['country']=='Greece'].births_2021.to_list()[0] 
print(f'{births_in_greece=}')
df2[(df2['births_2021'] >= 0.9 * births_in_greece) & (df2['births_2021'] <= 1.1 * births_in_greece)]
```
