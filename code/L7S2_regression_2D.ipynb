Regression example using artificially generated two dimensional data


```python
# use sklearn.dataset.make_regression to generate samples for a regression problem in 2D
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets

# create sample data 
X,y = datasets.make_regression(n_samples=100, n_features=1, noise=20, random_state=42)
X = np.interp(X, (X.min(), X.max()), (0,100))
y = np.interp(y, (y.min(), y.max()), (0,1000))

plt.rcParams["figure.figsize"] = 12,7
plt.plot(X,y,'.')
plt.title('Simulated data for regression');
```

```python
# plot the regression line
from sklearn.linear_model import LinearRegression

rgr = LinearRegression()
rgr.fit(X,y)
X2 = range(np.floor(X.min()).astype(int), np.ceil(X.max()).astype(int)+1)
y2 = rgr.coef_*X2 + rgr.intercept_
print(f'{rgr.coef_[0]}, {rgr.intercept_}')

plt.rcParams["figure.figsize"] = 12,7
plt.plot(X2, y2.T, color='red')
plt.plot(X,y,'.')
plt.title('Simulated data for regression & regression line')
```

```python
# plot points over the regression line
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
# compute linear regression line
rgr = LinearRegression()
rgr.fit(X,y)
print(f'y_hat = {rgr.coef_[0]} * x + {rgr.intercept_}')
# for each x compute y_hat
y_hat = np.array(X.T) * rgr.coef_ + rgr.intercept_
y_hat = y_hat.T
# plot graph
plt.rcParams["figure.figsize"] = 12,7
plt.plot(X, y_hat, color='grey')
plt.plot(X, y_hat, '.', color='red')
plt.plot(X, y,'.')
plt.title('Simulated data for regression & regression line');
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
r2 = r2_score(y, y_hat)
mae = mean_absolute_error(y, y_hat)
mse = mean_squared_error(y, y_hat)
rmse = np.sqrt(mse)
print(f'{r2=}, {mae=}, {mse=}, {rmse=}')
```
